```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('listings_initial_clean.csv')
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# Any duplicated data?

print(df.duplicated().value_counts())
```

```python
# What are the unique values in the columns?

unique_vals = df.nunique() # does not include NaN
print('Count of unique values in each column: \n{}'.format(unique_vals))
```

```python
graphs = [df['host_is_superhost', 'host_response_time']]

for i in graphs:
    i.value_counts().plot(kind='bar');
```
